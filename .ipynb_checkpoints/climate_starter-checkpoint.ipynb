# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
import datetime as dt
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
last_date = session.query(Measurement.date).order_by(Measurement.date.desc()).first()[0]
print(last_date, type(last_date))

last_datetime = dt.datetime.strptime(last_date, '%Y-%m-%d')
first_datetime = last_datetime - dt.timedelta(days=365)
print(first_datetime, last_datetime)
```

```python
prcp = (session.query(Measurement.date, Measurement.prcp)
                   .filter(Measurement.date > first_datetime)
                   .order_by(Measurement.date)
                   .all())

df = pd.DataFrame(prcp)
df = df.set_index('date').sort_index(ascending=True)
df.plot(rot=90,figsize=(10,5), title="Rainfall: 8/23/16 - 8/23/17")
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
station_sums = (session.query(Measurement.station, func.count(Measurement.prcp))
                       .group_by(Measurement.station)
                       .order_by(func.count(Measurement.station).desc())
                       .all())
station_sums
```

```python
temps = (session.query(func.min(Measurement.tobs), func.max(Measurement.tobs), func.avg(Measurement.tobs))
                        .filter(Measurement.station == station_sums[0][0])
                        #.order_by(func.count(Measurement.station).desc())
                        .all())
print(f"Most active station: {station_sums[0][0]}")
print(f"Lowest recorded temp: {temps[0][0]}")
print(f"Highest recorded temp: {temps[0][1]}")
print(f"Average recorded temp: {round(temps[0][2],1)}")
```

## Bonus Challenge Assignment

```python
# Identify the average temperature in June at all stations 
# across all available years in the dataset. Do the same for December temperature.
from sqlalchemy import extract


jun_temps = (session.query(Measurement.tobs)
                       .filter(extract('month', Measurement.date)==6)
                       .order_by(Measurement.date)
                       .all())

dec_temps = (session.query(Measurement.tobs)
                       .filter(extract('month', Measurement.date)==12)
                       .order_by(Measurement.date)
                       .all())
```

```python
round(np.array(jun_temps).mean(), 2), round(np.array(dec_temps).mean(), 2)
```

```python
def pvalue_eval(pval, a=.05):
    if pval<a :
        return f"{pval} is Statistically Significant (reject the null hypothesis)"
    return f"{pval} is Statistically Non-Significant (failed to reject the null hypothesis)"
```

```python
import scipy.stats as stats
ret = stats.ttest_ind(np.array(jun_temps), np.array(dec_temps), equal_var=False)
print(ret)
pvalue_eval(ret[1])
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
#print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
trip_start = '2017-08-20'
trip_end = '2017-08-28'

trip_temps = calc_temps(trip_start, trip_end)
trip_min, trip_avg, trip_max = round(trip_temps[0][0],2), round(trip_temps[0][1],2), round(trip_temps[0][2],2)
trip_min, trip_avg, trip_max
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)

error = trip_max - trip_min

plt.bar(f"{trip_start} \n to {trip_end}",trip_avg, yerr=error, color = 'orangered', alpha = .75)
plt.title('Vacation Average Temps')
plt.ylabel('Temp (F)')
plt.ylim(0, 100)
plt.xlim(-1, 1)

```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation
prev_year_start = '2016-08-20'
prev_year_end = '2016-08-28'

hist_rain = session.query(Station.station, Station.name, Station.latitude, 
                           Station.longitude, Station.elevation, func.avg(Measurement.prcp)).\
                            filter(Measurement.station == Station.station).\
                            filter(func.strftime("%Y-%m-%d", Measurement.date) >= prev_year_start).\
                            filter(func.strftime("%Y-%m-%d", Measurement.date) <= prev_year_end).\
                            group_by(Station.station).\
                            order_by(func.avg(Measurement.prcp).desc()).all()

df = pd.DataFrame(hist_rain, columns = ['Station ID', 'Station Name', 'Latitude', 'Longitude', 
                                                              'Elevation', 'Avg. Precipitation (in.)'])

df
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)

def daily_normals(date):
    """Daily Normals.
    
    Args:
        date (str): A date string in the format '%m-%d'
        
    Returns:
        A list of tuples containing the daily normals, tmin, tavg, and tmax
    
    """
    
    sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
    return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()
    
daily_normals("01-01")
```

```python
# calculate the daily normals for your trip
from datetime import timedelta, date

trip_dates = []

def daterange(date1, date2):
    for n in range(int ((date2 - date1).days)+1):
        yield date1 + timedelta(n)
        
for dt in daterange(date(2017, 8, 20), date(2017, 8, 28)):
    trip_dates.append(dt.strftime("%m-%d"))
    
daily_norms = []

for trip_date in trip_dates:
    normals = {}
    normals['Date'] = f"2019-{trip_date}"
    daily_norm = daily_normals(trip_date)
    normals['Min'] = daily_norm[0][0]
    normals['Avg'] = round(daily_norm[0][1],2)
    normals['Max'] = daily_norm[0][2]
    daily_norms.append(normals)
    
daily_norms


# push each tuple of calculations into a list called `normals`

# Set the start and end date of the trip

# Use the start and end date to create a range of dates

# Stip off the year and save a list of %m-%d strings

# Loop through the list of %m-%d strings and calculate the normals for each date
```
